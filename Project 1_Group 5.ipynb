```python
# Dependencies and Setup
import matplotlib as plt
import pandas as pd
import scipy.stats as st
```

```python
# Pricing data files
house_price_path = "resources/Housing_Price.csv"
rental_cost_path = "resources/Rental_Cost_Cleaned.csv"

# Read the housing price data
house_price_df = pd.read_csv(house_price_path)
house_price_df.head()
```

```python
# Read the rental cost data
rental_cost_df = pd.read_csv(rental_cost_path)
rental_cost_df.head()
```

```python
#Delete extraneous column in house price ( 'Locality0' column)
del house_price_df["Locality 0"]
house_price_df.head()
```

```python
#Delete extraneous column in rental cost ( all 'unnamed'columns)
rental_cost_df=rental_cost_df.drop(rental_cost_df.columns[rental_cost_df.columns.str.contains('Unnamed',case=False)],axis=1)

#rental_cost_df=rental_cost_df.drop(rental_cost_df.index[0])
rental_cost_df=rental_cost_df.drop(labels=0,axis=0)

#Rename columns to be consistent with Housing Price headers
rental_cost_df=rental_cost_df.rename(columns={"Jun-10":"2010",
                                            "Jun-11":"2011",
                                            "Jun-12":"2012",
                                            "Jun-13":"2013",
                                            "Jun-14":"2014",
                                            "Jun-15":"2015",
                                            "Jun-16":"2016",
                                            "Jun-17":"2017",
                                            "Jun-18":"2018",
                                            "Jun-19":"2019",
                                            "Jun-20":"2020",
                                            })
rental_cost_df.head()
```

```python
#Split column on delimter -
#rental_cost_df=rental_cost_df['Locality'].str.split('-',expand=True)
#rental_cost_df.head()
df1=house_price_df
df2=rental_cost_df

new_df2=df2['Locality'].str.split('-', expand=True).stack().reset_index(level=0).set_index('level_0').rename(columns={0:'Locality'}).join(df2.drop('Locality', axis=1))
new_df2
```

```python
#Filter House Price dataframe by Locality in Rental Cost dataframe
#df1=house_price_df
#df2=rental_cost_df

merge_df=pd.merge(df1,new_df2,on="Locality")
merge_df

```
